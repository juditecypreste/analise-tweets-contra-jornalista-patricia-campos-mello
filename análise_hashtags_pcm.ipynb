```python
# Importação de bibliotecas e módulos
import os
from pathlib import Path
import pandas as pd
```

```python
# Define função de checagem de df
def check_df(df):
    """ exibe primeiras linhas do dataframe e suas quantidades de linhas e colunas.
    INPUT:
    df - o dataframe analisado
    OUTPUT:
    exibe as 3 primeiras linhas do dataframe e imprime:
    "O dataframe possui {} linhas e {} colunas", com os totais de linhas e colunas.    
    """
    display(df.head(3))
    display(df.tail(3))
    print("O dataframe possui {} linhas e {} colunas".format(df.shape[0],df.shape[1]))
```

```python
# Me mostre tudo
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Abrindo a pasta com os csvs
bases = Path("bases_camposmello/")
```

```python
# Juntando todos os dfs
df_total = pd.DataFrame()
dfs = []

for arquivos in os.listdir(bases):
    print(arquivos)
    dfs.append(pd.read_csv(f"{bases}/{arquivos}"))

df_total = pd.concat(dfs)
```

```python
# Verificando as urls dos tweets 
for index, row in df_total.iterrows():
        print(row.link)
```

```python
# Conferindo o df
check_df(df_total)
```
